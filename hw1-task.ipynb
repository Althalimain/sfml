```python
# (6) Создать новый строковый столбец в data frame-е, который:



# принимает значение A, если значение LIMIT_BAL <=10000
# принимает значение B, если значение LIMIT_BAL <=100000 и >10000
# принимает значение C, если значение LIMIT_BAL <=200000 и >100000
# принимает значение D, если значение LIMIT_BAL <=400000 и >200000
# принимает значение E, если значение LIMIT_BAL <=700000 и >400000
# принимает значение F, если значение LIMIT_BAL >700000

for i in range(len(data_csv)):
    if data_csv.at[i, 'LIMIT_BAL'] <= 10000:
        data_csv.at[i, 'IDX_LIMIT'] = 'A'

    elif (data_csv.at[i,'LIMIT_BAL'] <= 100000) & (data_csv.at[i,'LIMIT_BAL'] > 10000):
        data_csv.at[i, 'IDX_LIMIT'] = 'B' 
        
    elif (data_csv.at[i,'LIMIT_BAL'] <= 200000) & (data_csv.at[i,'LIMIT_BAL'] > 100000):
        data_csv.at[i, 'IDX_LIMIT'] = 'C'   
        
    elif (data_csv.at[i,'LIMIT_BAL'] <= 400000) & (data_csv.at[i,'LIMIT_BAL'] > 200000):
        data_csv.at[i, 'IDX_LIMIT'] = 'D'   
        
    elif (data_csv.at[i,'LIMIT_BAL'] <= 700000) & (data_csv.at[i,'LIMIT_BAL'] > 400000):
        data_csv.at[i, 'IDX_LIMIT'] = 'E'   
        
    elif data_csv.at[i,'LIMIT_BAL'] > 700000:
        data_csv.at[i, 'IDX_LIMIT'] = 'F' 


data_csv.head()
#TODO
```

```python
def fSort(inum):
    outS = ''
    if inum <= 10000:
        outS = 'A'
    elif (inum <= 100000) & (inum > 10000):
        outS = 'B'
    elif (inum <= 200000) & (inum > 100000):
        outS = 'C'
    elif (inum <= 400000) & (inum > 200000):
        outS = 'D'
    elif (inum <= 700000) & (inum > 400000):
        outS = 'E'
        
    elif inum > 700000:
        outS = 'F'
   
    return outS
    
```

```python
data_csv['IDX_LIMIT2'] = data_csv['LIMIT_BAL'].map(fSort)
data_csv.head()
```
